```python
image_dir = input("/content/SUBBED (default: '/workspace/data/me'): ") or '/workspace/data/me'
class_tokens = input("GGRRLL (default: 'shs man'): ") or 'shs man'

config_dataset = f"""
[[datasets]]
keep_tokens = 2

  [[datasets.subsets]]
  image_dir = '{image_dir}'
  class_tokens = '{class_tokens}'
  num_repeats = 10

[general]
resolution = 1024
"""

with open('config_dataset.toml', 'w') as f:
    f.write(config_dataset)
```

```python
pretrained_model_name_or_path = input("Enter the path of the pretrained model (default: './models/sd_xl_base_0.9.safetensors'): ") or './models/sd_xl_base_0.9.safetensors'
learning_rate = float(input("Enter the learning rate (default: 5e-5): ") or 5e-5)
output_dir = input("Enter the output directory (default: 'ComfyUI/models/checkpoints/'): ") or 'ComfyUI/models/checkpoints/'
gradient_accumulation_steps = int(input("Enter the gradient accumulation steps (default: 2): ") or 2)
sample_prompts = input("Enter the sample prompts (default: 'prompts.txt'): ") or 'prompts.txt'
train_batch_size = int(input("Enter the train batch size (default: 1): ") or 1)
max_train_steps = int(input("Enter the max train steps (default: 1000): ") or 1000)

config = f"""
[model_arguments]
v2 = false
v_parameterization = false
pretrained_model_name_or_path = "{pretrained_model_name_or_path}"

[optimizer_arguments]
optimizer_type = "adafactor"
optimizer_args = [ "scale_parameter=False", "relative_step=False", "warmup_init=False" ]
lr_scheduler = "cosine_with_restarts"
lr_warmup_steps = 100
learning_rate = {learning_rate}
max_grad_norm = 0.0
train_text_encoder = true

[dataset_arguments]
debug_dataset = false

[training_arguments]
output_dir = "{output_dir}"
output_name = "1000_lr1e-5_cosine"
save_precision = "fp32"
save_n_epoch_ratio = 1
save_state = false
train_batch_size = {train_batch_size}
mem_eff_attn = false
max_train_steps = {max_train_steps}
max_data_loader_n_workers = 1
persistent_data_loader_workers = true
gradient_checkpointing = true
gradient_accumulation_steps = {gradient_accumulation_steps}
mixed_precision = "bf16"
logging_dir = "logs"
log_prefix = "last"
sample_prompts = "{sample_prompts}"
log_with = "wandb"
xformers = true
cache_latents = true
full_bf16 = true

[sample_prompt_arguments]
sample_every_n_steps = 10000
sample_sampler = "euler a"

[saving_arguments]
save_model_as = "safetensors"
"""

with open('config.toml', 'w') as f:
    f.write(config)
```
